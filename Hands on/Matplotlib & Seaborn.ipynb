```python
years=[1960,1970,1980,1990,2000,2010]
Population=[21.91,23.90,24.80,20.93,22.30,26.90]
```

```python
#line plot 
plt.plot(years,Population)
plt.show()
```

```python
plt.scatter(years,Population)
plt.show()
```

```python
colors=['violet','tomato','crimson','green','blue','pink']
plt.scatter(years,Population,c=colors)
plt.show()
```

```python
years=[1960,1970,1980,1990,2000,2010]
Canada_Pop=[21.91,23.90,24.80,20.93,22.30,26.90]
USA_Pop=[32.91,24.80,23.60,21.92,32.30,26.90]
plt.plot(years,Canada_Pop,marker='+',mew=8,ls='-')
plt.plot(years,USA_Pop,ls='--',lw=1)
plt.legend(['Canada','USA'],loc='best')
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(years,Canada_Pop,marker='+',mew=8,ls='-')
plt.title('Population of Canada')
plt.subplot(1,2,2)
plt.plot(years,USA_Pop,ls='--',lw=1)
plt.title('Population of United States')
plt.show()
```

```python
import pandas as pd 
import seaborn as sns
```

```python
#Strip plot
data1=pd.read_csv('ozone1.csv')
```

```python
sns.stripplot(x='State Name',y='Site Num',data=data1,size=10,jitter=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x='State Name',y='Site Num', data=data1)
plt.show()
```

```python
sns.jointplot(x='State Name',y='Site Num',data=data1,kind='scatter')
plt.show()
```

```python
sns.pairplot(data=data1, vars=['State Name', 'Site Num'], kind='scatter')
plt.show()
```
